```python
# Importing Libraries
import selenium
import pandas as pd
import time
import requests
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver
```

```python
# Activating the chrome driver
driver = webdriver.Chrome(r"F:\Saiprakash\Flip Robo\WS_Assignments & Solutions\chromedriver.exe")
```

```python
# Loading the website
driver.get('https://www.amazon.in/')
```

```python
search_bar = driver.find_element_by_id("twotabsearchtextbox")    # Locating search bar by id
search_bar.clear()                                               # clearing search bar if any item is searched
# Giving the input of the desired item
inp = input('Enter the clothing category you want to search : ')
search_bar.send_keys(inp)                                   # sending user input to search bar
```

```python
search_button = driver.find_element_by_xpath('//div[@class="nav-search-submit nav-sprite"]/span/input')       # Locating search button by xpath
search_button.click()                                                                # Clicking the search button to start search
```

```python
Saree_images = driver.find_elements_by_xpath('//img[@class="s-image"]')
```

```python
img_urls = []
img_data = []
for page in range(0,4):
    for image in Saree_images:
        source= image.get_attribute('src')
        if source is not None:
            if(source[0:4] == 'http'):
                img_urls.append(source)
len(img_urls)
```

```python
for i in range(len(img_urls[:240])):
    if i >= len(img_urls[:240]):
        break
    print("Downloading {0} of {1} images" .format(i, len(img_urls[:240])))
    response= requests.get(img_urls[i])
    file = open(r"F:\Saiprakash\Flip Robo\Image Scraping and Classification\Saree Images\img"+str(i)+".jpg", "wb")
    file.write(response.content)
```
