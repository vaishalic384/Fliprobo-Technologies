```python
#2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of
#release) and make data frame
import pandas as pd
import numpy as np

import requests
from requests import get
from bs4 import BeautifulSoup

# Creating the lists we want to write into
titles = []
years = []
imdb_ratings = []

headers = {'Accept-Language': 'en-US, en;q=0.5'}
pages = np.arange(1, 101, 50)
pages

for page in pages:
    # Getting the contents from the each url
    page = requests.get('https://www.imdb.com/search/title/?groups=top_1000&start=' + str(page) + '&ref_=adv_nxt', headers=headers)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    movie_div = soup.find_all('div', class_='lister-item mode-advanced')
    
    
    for container in movie_div:
        # Scraping the movie's name
        name = container.h3.a.text
        titles.append(name)
        
        # Scraping the movie's year
        year = container.h3.find('span', class_='lister-item-year').text
        years.append(year)
        
             
        # Scraping the rating
        imdb = float(container.strong.text)
        imdb_ratings.append(imdb)
        
        
        
movies = pd.DataFrame({'movie':titles,
                       'year':years,
                       'imdb_rating':imdb_ratings
                       })

movies
```

```python
#3.Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year
#of release) and make data frame.


import pandas as pd
import numpy as np

import requests
from requests import get
from bs4 import BeautifulSoup

# Creating the lists we want to write into
titles = []
years = []
imdb_ratings = []


page = requests.get('https://www.imdb.com/list/ls009997493/')
soup = BeautifulSoup(page.text, 'html.parser')

movie_div=soup.find_all('div',class_='lister-item-content')
#print(movie_div)
for i in movie_div:
     # Scraping the movie's name
    name=i.find('h3',class_='lister-item-header').text.replace(' ','').split()[-2]   
    titles.append(name)
    
    # Scraping the movie's year
    year=i.find('h3',class_='lister-item-header').text.replace(' ','').split()[-1]
    years.append(year)
    
    # Scraping the rating
    rating=i.find('span',class_="ipl-rating-star__rating").text
    imdb_ratings.append(rating)
        
        
movies = pd.DataFrame({'movie':titles,
                       'year':years,
                        'imdb_rating':imdb_ratings
                      })

movies
```

```python
#4.Write a python program to scrap book name, author name, genre and book review of any 5 books from
#‘www.bookpage.com
import pandas as pd
import numpy as np

import requests
from requests import get
from bs4 import BeautifulSoup


page=requests.get('https://bookpage.com/reviews')
soup=BeautifulSoup(page.text, 'html.parser')
book_div=soup.find_all('div',class_='flex-article-content')
#print(book_div)
count=1
while count<5:
    for i in book_div:
  
        #scraping book name
        name=i.find('h4',class_='italic').text
        print("\nName of the book:",name)
    
        #scraping author name
        auth_name=i.find('p',class_='sans bold').text
        print("Author of the book:",auth_name)
    
        #scraping genre
        genre_link=i.find('p',class_='genre-links hidden-phone')
        genre=genre_link.find('a').text
        print("Genre of book:",genre)
        #genre=genre_link.a['href'].text
    
        #scraping the review
        review=i.find('p',class_='excerpt').text
        print("\nreview:",review)
        print("------------------------------------------------------------------------------------------------------")
        count=count+1
        if count==6:
            break
            
```

```python
#8.Write a python program to extract information about the local weather from the National Weather Service 
#website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day 
#extended forecast display for the city. The data should include period, short description, temperature and 
#description.
import pandas as pd

period_short_des=[]
temperature=[]
temp_desc=[]

page=requests.get('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YNGxuBHivIU')
soup=BeautifulSoup(page.text,'html.parser')

weather_div=soup.find_all('ul',id="seven-day-forecast-list")
#print(weather_div)
for i in weather_div:
    container=i.find_all('div',class_='tombstone-container')
    for j in container:
        desc=j.find('p',class_='period-name').text
        period_short_des.append(desc)
    
        temp=j.find('p',class_='temp temp-low')
        temp1=j.find('p',class_='temp temp-high')
        if temp:
            temperature.append(temp.text)
        else:   
            temperature.append(temp1.text)
    
        w_desc=j.find('p',class_='short-desc').text
        temp_desc.append(w_desc)
    
    
weather_info = pd.DataFrame({'Period':period_short_des,
                          'Temperature':temperature,
                           'Weather Short Description':temp_desc
                            
                      })

weather_info
```

```python
#9.Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, 
#company name, CTC, and apply date.

import pandas as pd
job_title=[]
company_name=[]
CTC=[]
apply_date=[]

page=requests.get('https://internshala.com/fresher-jobs')
soup=BeautifulSoup(page.text,'html.parser')

job_div=soup.find_all('div',class_='internship_meta')
for i in job_div:
    title=i.find('div',class_='heading_4_5 profile').text.replace('\n','')
    job_title.append(title)
    
    company=i.find('div',class_='heading_6 company_name').text.replace('\n','')
    company_name.append(company)
    
    
    
    other_detail=i.find('div',class_='other_detail_item_row').text.replace('\n','')
    apply_date.append(other_detail)
        
       
    other_detail1=i.find('div',class_='other_detail_item').text.replace('\n','')
    CTC.append(other_detail1)
    

    
job_info = pd.DataFrame({'Job Title':job_title,
                        'Company Name':company_name,
                         'CTC':CTC,
                        'Apply Date':apply_date
                            
                      })

job_info
```
