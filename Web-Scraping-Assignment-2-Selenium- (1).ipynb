```python
#Q1: Write a python program to scrape data for “Data Analyst” Job position in 
#“Bangalore” location. You have to scrape the job-title, job-location, company_name,
#experience_required. You have to scrape first 10 jobs data

path='chromedriver.exe'
driver=webdriver.Chrome(path)
driver.get('https://www.naukri.com/')

#writing 'Data Analyst' in search
search=driver.find_element_by_id('qsb-keyword-sugg')
search.send_keys('Data Analyst')

#writing location as 'Banglore'
search=driver.find_element_by_id('qsb-location-sugg')
search.send_keys('Bangalore')

#clicking on search button
search=driver.find_element_by_class_name('search-btn')
search.click()

```

```python
#creating empty list for storing data
job_location=[]
job_titles=[]
company_names=[]
exp_req=[]
```

```python
#scraping data for job titles
titles = driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")

for i in titles:
    job_titles.append(i.text)
```

```python
#scraping data for comapany names.

comapany=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
for i in comapany:
    company_names.append(i.text)
```

```python
#scraping data for job location


location=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']//span")
for i in location:
    job_location.append(i.text)
```

```python
#scraping data for experience required.

exp=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']//span")
for i in exp:
    exp_req.append(i.text)
```

```python
#checking length of each list
print(len(job_titles),len(job_location),len(company_names),len(exp_req))


#storing data in a dataframe and dispalying only 10 rows
jobs=pd.DataFrame({})
jobs['Job Title']=job_titles
jobs['Job Location']=job_location
jobs['Company Name']=company_names
jobs['Experience Required']=exp_req
jobs[:10]

```

```python
driver.close()
```
